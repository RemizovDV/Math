### –†–µ–º–∏–∑–æ–≤ –î–º–∏—Ç—Ä–∏–π –í–ª–∞–¥–∏–º–∏—Ä–æ–≤–∏—á

```python
from math import factorial
from math import erf
from math import sqrt
import numpy as np
```

```python
def c(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
def Bernul(p,n,k):
    return c(n,k)*(p**k)*((1-p)**(n-k))
```

```python
def Puasson(lambda_,m):
    return (lambda_**m / np.math.factorial(m)) * np.exp(-lambda_)
```

```python
def GaussFR(MathExp,std,x):
    return (1+erf((x-MathExp)/(sqrt(2*std*std))))/2
```

**1.** –°–ª—É—á–∞–π–Ω–∞—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ A –∏–º–µ–µ—Ç —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ (200, 800]. 
–ù–∞–π–¥–∏—Ç–µ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é.


$$f(x) = \begin{cases} 0, –µ—Å–ª–∏ \; x \le a; \\ \frac{1}{b - a}, –µ—Å–ª–∏ \; a < x \le b; \\ 0, \; –µ—Å–ª–∏ \; x > b. \end{cases}$$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ  ùëÄ(ùëã)  —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

$$M(X) = \frac{a + b}{2}$$

```python
print(f'–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ {(200+800)/2}')
```

–ü–æ —Å–ª–µ–¥—É—é—â–µ–π —Ñ–æ—Ä–º—É–ª–µ –º–æ–∂–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏—é $D(X)$:

$$D(X) = \frac{(b - a)^2}{12}$$

**2.** –û —Å–ª—É—á–∞–π–Ω–æ–π –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–µ B –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –µ–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞–≤–Ω–∞ 0.2.
–ú–æ–∂–Ω–æ –ª–∏ –Ω–∞–π—Ç–∏ –ø—Ä–∞–≤—É—é –≥—Ä–∞–Ω–∏—Ü—É –≤–µ–ª–∏—á–∏–Ω—ã B –∏ –µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–Ω–∞—è, —á—Ç–æ –ª–µ–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ —Ä–∞–≤–Ω–∞ 0.5?
–ï—Å–ª–∏ –¥–∞, –Ω–∞–π–¥–∏—Ç–µ –µ–µ.


$$D(X) = \frac{(b - a)^2}{12}$$

–ü–æ–¥—Å—Ç–∞–≤–∏–º –≤ –¥–∞–Ω–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ –Ω–∞–π–¥–µ–º –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
b=np.linspace(-3,3,600)
plt.plot(b,(b-0.5)**2-12*0.2)
plt.xlabel('b')
plt.ylabel('f(b)')
plt.grid(True)
plt.gca().set_aspect('equal')
```

```python
from scipy.optimize import fsolve

def eq(b):
    return ((b-0.5)**2-12*0.2)

b1=fsolve(eq,(1))
print (f'–ü–µ—Ä–≤–æ–µ —Ä–µ—à–µ–Ω–∏–µ b= {b1}')
```

```python
b2=fsolve(eq,(-1))
print (f'–≤—Ç–æ—Ä–æ–µ  —Ä–µ—à–µ–Ω–∏–µ b={b2}')
```

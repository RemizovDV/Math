4.1. Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
x=np.linspace(-3,3,300)
plt.plot(x,(np.exp(x)-1)/x+1)
plt.plot(x,x*x-1)
plt.xlabel('x')
plt.xlabel('y')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def eq(point):
    x,y =point
    return (y-x*x+1, np.exp(x)+x-x*y-1)

x1,y1=fsolve(eq,(-1.5,1.8))
print ('Первое решение',x1,y1)
```

```python
x2,y2=fsolve(eq,(2.5,6))
print ('Второе решение',x2,y2)
```

4.2. Решите систему уравнений и неравенств:

    y = x2 – 1

    exp(x) + x∙(1 – y) > 1


подставим y=x^2-1 в неравентсво
exp(x)+x*(2-x^2)-1>0

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
x=np.linspace(-3,3,300)
plt.plot(x,np.exp(x)+x*(2-x*x)-1)
plt.xlabel('x')
plt.xlabel('f(x)')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve

def eq(x):
    return (np.exp(x)+x*(2-x*x)-1)

x1=fsolve(eq,(-1.5))
print ('Первое решение x<',x1, 'y=x^2-1')
```

```python
x2=fsolve(eq,(0.1))
x3=fsolve(eq,(2.5))
print (x2,'<x<',x3, 'y=x^2-1')
```

Так как решение выполнялось численно, то результат содержит погрешность вычисления

Напишите код, проверяющий любую из теорем сложения или умножения вероятности на примере подбрасывания монетки.

```python
import numpy as np
k,m,l=0,0,0
n=1000
for i in range(0, n):
    x=np.random.randint(2)
    y=np.random.randint(2)
    # 0 - выпадение решки, 1 - выпадение орла
    # x,y - монета 1 и 2 соответственно
    if x==0:
        k+=1
    if y==0:
        m+=1
    if (x==0) and (y==0):
        l+=1
print('частота выпадения решки на первой монете', k/n)
print('частота выпадения решки на второй монете', m/n)
print('частота выпадения решки на обоих монетах', l/n)
```
